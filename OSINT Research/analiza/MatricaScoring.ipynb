```python
import pandas as pd
import os

# Matrica finale sipas tabelave te ristrukturuara
data = {
    'Profil': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6'],
    
    'V1_EXIF (25)': [25, 8, 0, 0, 0, 0],
    # P1: 10/10 GPS sakte(+15)+Model(+5)+Timestamp(+5)=25
    # P2: 2/10 GPS pjeserisht(+5)+Model(+3)+Timestamp(0)=8
    # P3-P5: Zero EXIF=0
    # P6: GPS Rome, ndaj P1-P5=0
    
    'V2_CrossPlatform (20)': [14, 10, 2, 0, 0, 0],
    # P1: 5 hits pertej baseline + variante username=14
    # P2: 6 hits por username te ndryshme=10
    # P3: 2 hits (vetem user_4729)=2
    # P4-P6: Zero pertej baseline=0
    
    'V3_Stilometri (20)': [20, 18, 12, 8, 3, 0],
    # P1: Sakte te dyja + 8 platforma (corpus i bollshem)=20
    # P2: Sakte te dyja + 8 platforma variante (corpus i forte)=18
    # P3: Sakte te dyja + 5 platforma username unik (corpus mesatar)=12
    # P4: Sakte vetem me 1 metode (modifikim aktiv)=8
    # P5: Sakte por corpus i paaksessueshem (1 platforme, llogari e re)=3
    # P6: Gjuhe komplet ndryshe=0
    
    'V4_Orare (15)': [15, 13, 14, 8, 2, 0],
    # P1: 98.7% dite, pa VPN=15
    # P2: 97.8% dite, VPN 25% (noise minimal)=13
    # P3: 100% dite, VPN 93% (pa efekt real)=14
    # P4: 87.8% dite (i mundshem por jo i sigurte)=8
    # P5: 61.9% dite (i painferencueshem)=2
    # P6: Ndaj P1-P5=0
    
    'V5_Rrjeti (10)': [10, 8, 3, 0, 0, 0],
    # P1: 10 bridge cross-platform (5+ nyje)=10
    # P2: 8 bridge (5+ nyje)=8
    # P3: 2 bridge te dobeta (1-2 nyje)=3
    # P4: 0 bridge=0
    # P5: 0 bridge (3 lidhje totale)=0
    # P6: Ndaj P1-P5=0
    
    'V6_Breach (10)': [0, 0, 0, 0, 0, 0],
    # Te gjitha 0 ne testim real (email fiktive @example.com)
    # HIBP ktheu "No pwnage found" per te gjitha
    # Vlera teorike dokumentohet ne Kapitullin 6
}

df = pd.DataFrame(data)
df['TOTAL'] = df[['V1_EXIF (25)', 'V2_CrossPlatform (20)', 'V3_Stilometri (20)', 
                   'V4_Orare (15)', 'V5_Rrjeti (10)', 'V6_Breach (10)']].sum(axis=1)

def nivel(score):
    if score >= 80: return "V - Konfirmim i plote"
    elif score >= 60: return "IV - Identifikim i forte"
    elif score >= 40: return "III - Indikatore mesatare"
    elif score >= 20: return "II - Sinjale te dobeta"
    else: return "I - De-anonimizim i pamundshem"

df['Niveli'] = df['TOTAL'].apply(nivel)

print("=" * 95)
print("MATRICA E SCORING-UT FINALE")
print("=" * 95)
print(df.to_string(index=False))
print("=" * 95)

print("\nJUSTIFIKIMI PER CDO PROFIL:")
print("-" * 50)
justifikime = {
    'P1': 'Zero mbrojtje. Te gjitha vektoret pozitive. EXIF i plote, username identik, stil i dallueshem, UTC+1 i qarte, rrjet i centralizuar.',
    'P2': 'Mbrojtje bazike. EXIF pjeserisht i fshire, username variante, stil i dallueshem, VPN pa efekt ne orare.',
    'P3': 'Mbrojtje mesatare. EXIF i fshire, username unik, stil neutral por me zakone, VPN nuk fsheh oraret, 2 bridge te dobeta.',
    'P4': 'Mbrojtje avancuare. Zero EXIF, zero cross-platform, stil i modifikuar (mashtron 1 metode), orare kaotike, rrjete te izoluara.',
    'P5': 'Mbrojtje maksimale. Zero ne te gjitha vektoret perveq stilometrise minimale (corpus i paaksessueshem).',
    'P6': 'Kontroll. Zero lidhje me P1-P5 ne cdo vektor. Framework-u nuk prodhon false positives.',
}
for p, j in justifikime.items():
    print(f"  {p}: {j}")

base = r"C:\Users\Laura Biba\Desktop\OSINT Research"
df.to_csv(os.path.join(base, 'rezultate', 'matrica_scoring_final.csv'), index=False)
print(f"\nRezultati i ruajtur ne rezultate/matrica_scoring_final.csv")
```
