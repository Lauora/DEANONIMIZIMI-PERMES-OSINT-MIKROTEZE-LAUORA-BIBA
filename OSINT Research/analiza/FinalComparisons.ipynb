```python
import matplotlib.pyplot as plt
import numpy as np
import os

base = r"C:\Users\Laura Biba\Desktop\OSINT Research"

# ============================================
# FIGURA 1: Score i Pritur vs Score Real
# ============================================
fig, ax = plt.subplots(figsize=(12, 6))

profiles = ['P1\n(Zero)', 'P2\n(Bazike)', 'P3\n(Mesatare)', 'P4\n(Avancuare)', 'P5\n(Maksimale)', 'P6\n(Kontroll)']
pritur_min = [90, 45, 17, 7, 2, 0]
pritur_max = [100, 60, 28, 12, 5, 0]
real = [84, 57, 31, 16, 5, 0]

x = np.arange(len(profiles))
width = 0.25

# Range i pritur si shirit me transparence
for i in range(len(profiles)):
    ax.bar(x[i] - width/2, pritur_max[i], width, color='#3498DB', alpha=0.3, 
           label='Range i pritur' if i == 0 else '')
    ax.bar(x[i] - width/2, pritur_min[i], width, color='white', edgecolor='#3498DB', linewidth=1.5)

# Score real
bars = ax.bar(x + width/2, real, width, color='#E74C3C', label='Score real', edgecolor='black')

# Shto vlerat mbi bars
for i, v in enumerate(real):
    ax.text(x[i] + width/2, v + 1.5, str(v), ha='center', fontweight='bold', fontsize=12)
    ax.text(x[i] - width/2, pritur_max[i] + 1.5, f'{pritur_min[i]}-{pritur_max[i]}', 
            ha='center', fontsize=9, color='#3498DB')

# Diferenca
for i in range(len(profiles)):
    mid = (pritur_min[i] + pritur_max[i]) / 2
    diff = real[i] - mid
    sign = '+' if diff > 0 else ''
    if diff != 0:
        ax.annotate(f'{sign}{diff:.0f}', xy=(x[i], max(real[i], pritur_max[i]) + 6),
                   ha='center', fontsize=9, color='green' if abs(diff) <= 5 else 'orange',
                   fontweight='bold')

ax.set_xlabel('Profili', fontsize=12)
ax.set_ylabel('Score (0-100)', fontsize=12)
ax.set_title('Krahasimi: Score i Pritur vs Score Real', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(profiles)
ax.legend(loc='upper right')
ax.set_ylim(0, 115)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(base, 'figurat', 'pritur_vs_real.png'), dpi=150, bbox_inches='tight')
plt.show()
print("Figura 1 e ruajtur")
```
